# 🌋 Seismic Signal Analysis with Python

This notebook simulates and visualizes seismic signals, reads real miniSEED data, filters the signal, performs basic statistical analysis, and sends results to WhatsApp bot using webhook.

```python
# 2. Import libraries
import numpy as np
import matplotlib.pyplot as plt
from obspy import read
import pandas as pd
import requests
from scipy import signal
print("libraries imported successfully.")
```

```python
# 3. Load real data
st = read('../data/example.mseed')  # Change path accordingly
st.plot()
```

```python
# 4. Filter signal and plot
tr = st[0]
tr_filtered = tr.copy()
tr_filtered.filter("bandpass", freqmin=1, freqmax=10)

fig, axes = plt.subplots(2, 1, figsize=(12, 6))
axes[0].plot(tr.times(), tr.data, label='Original')
axes[0].legend()
axes[1].plot(tr_filtered.times(), tr_filtered.data, label='Filtered')
axes[1].legend()
plt.show()
```

```python
# 5. Statistik
stats = {
    'Mean': np.mean(tr_filtered.data),
    'Std Dev': np.std(tr_filtered.data),
    'Max': np.max(tr_filtered.data),
    'Min': np.min(tr_filtered.data)
}
df_stats = pd.DataFrame([stats])
df_stats
```

```python
# 6. Kirim ke WhatsApp Bot (via webhook)
webhook_url = "http://localhost:3000/incoming"  # Ganti jika perlu
data = {
    "from": "GeoLab",
    "message": f"Seismic Stats:\nMean: {stats['Mean']:.2f}\nStd Dev: {stats['Std Dev']:.2f}"
}
response = requests.post(webhook_url, json=data)
print(f"Webhook sent. Status code: {response.status_code}")
```
